Ahora, como funciona esto, piensa en como llenar una tabla donde las columnas y filas tienen que tener la misma cantidad de marcas que las filas y columnas vecinas, primero se marca una casilla, cualquier casilla, luego buscas donde falten mas marcas y repites el proceso hasta llenar los espacios que se ocupen, y es asi como funciona basicamente.

Esta parte busca todas las columnas y filas a las que les hace falta casillas marcadas:

<code>Xplaces=np.where(np.isin(xArray,[minX]))
Yplaces=np.where(np.isin(yArray,[minY]))</code>

Ahora de entre las secciones mas necesitadas, selecciona una al azar: 
    
<code>xC=Xplaces[0][r.randint(0,len(Xplaces[0])-1)]
yC=Yplaces[0][r.randint(0,len(Yplaces[0])-1)]</code>

Marca la casilla:

<code>matrix[xC,yC]=1

yArray[yC]=yArray[yC]+1</code>

Actualiza que columna y fila nesesitará mas:

<code>minX=min(xArray)
minY=min(yArray)</code>

## Potenciales errores
el programa es bastane similar al proceso que haria una persona pero como toda persona, con las condiciones adecuadas se puede equivocar, en el caso de este programa va a retornar la matriz en la que se trabó, la forma en la que el progama entra en un posible ciclo infinito es si <code>matrix[xC,yC]!=0</code> lo que significa que la casilla que se debe marcar, ya esta marcada, lo que significa que en alguna parte en el llenado de la matriz se marcó la casilla equivocada semejante a cuando uno se equivocaen las torres de hanoi solo que es muy dificil de identificar y revertir lo que deja solo dos opciones: 
<ul><li>Reiniciar el proceso</li><li>Tratar de arreglar la matriz</li></ul>
